```python
erstes_ausbildungsjahr = df[(df.entgelt < 2000) & (df.ausbildungsjahr == '1.')].sort_values(by='entgelt')
zweites_ausbildungsjahr = df[(df.entgelt < 2000) & (df.ausbildungsjahr == '2.')].sort_values(by='entgelt')
drittes_ausbildungsjahr = df[(df.entgelt < 2000) & (df.ausbildungsjahr == '3.')].sort_values(by='entgelt')

erstes_ausbildungsjahr.plot.scatter(y='branche', x='entgelt')
zweites_ausbildungsjahr.plot.scatter(y='branche', x='entgelt')
drittes_ausbildungsjahr.plot.scatter(y='branche', x='entgelt')
```
